```python
import requests
from google.colab import userdata
demo=userdata.get('alphavantage')
# replace the "demo" apikey below with your own key from https://www.alphavantage.co/support/#api-key
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_WEEKLY_ADJUSTED&symbol=CIM&apikey='+demo
r = requests.get(url)
data = r.json()
print(data)
```
